****ï¼ŠWhat is the SparkContext:*****
Main entry point for Spark functionality.
A SparkContext represents the connection to a Spark cluster, and can be used to create RDDs, accumulators and broadcast variables on that cluster.
Only one SparkContext may be active per JVM.
You must stop() the active SparkContext before creating a new one.

```python
from pyspark import SparkContext,SparkConf
```

```python
####First Way to create Spark Context##
conf = SparkConf().setAppName("Spark_Demo").setMaster("local")
sc = SparkContext(conf=conf)
```

```python
sc.getConf().getAll()
```

```python
sc.stop()
```

RDD can be created in two ways: by loading an external dataset, or by transforming one RDD into another.
